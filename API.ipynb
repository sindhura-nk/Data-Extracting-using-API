## SBI India BSE data 

```python
import requests

url = "https://alpha-vantage.p.rapidapi.com/query"

querystring = {"function":"TIME_SERIES_DAILY","symbol":"SBIN.BSE","outputsize":"compact","datatype":"json"}

headers = {
	"x-rapidapi-key": "9747ecd9e1mshe3c551622099032p1879f6jsn07a8de6306dc",
	"x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

data =response.json()
data
```

```python
data.keys()
```

```python
daily = data['Time Series (Daily)']
daily
```

```python
import pandas as pd
df = pd.DataFrame(daily).T
df.head()
```

```python
df.info()
```

By default, pandas considers date as string. Change date index to datetime

```python
df.index = pd.to_datetime(df.index)
```

By default, pandas considers values as strings. Let's convert all the columnar values to float.

```python
# we need to convert the values present in the columns. so we are using for loop. 
# If we apply df.columns.astype it will try to convert column names,which is not what we want
for i in df.columns:
    df[i] = df[i].astype(float)
```

```python
df.info()
```

Sort the dates from oldest to latest

```python
df = df.sort_index(ascending=True)
df
```

Save the data into csv file

```python
df.index = df.index.set_names(['Date'])
```

```python
df.to_csv('SBI BSE STOCK PRICE.csv',index_label='Date')
```
